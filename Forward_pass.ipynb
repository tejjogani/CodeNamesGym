```python
all_words = random.sample(model.wv.index_to_key, 16)
red_words = all_words[0:7]
bad_words = all_words[7:]
```

```python
R = defaultdict(list)
```

```python
from scipy import spatial
```

```python
model.wv["adopted"]
```

```python
for red_word in red_words:
    for word in words:
        try:
            R[red_word].extend([spatial.distance.cosine(model.wv[word], model.wv[red_word])])
        except:
            R[red_word].extend([float("inf")])
```

```python
B = defaultdict(list)
```

```python
for bad_word in bad_words:
    for word in words:
        try:
            B[bad_word].extend([spatial.distance.cosine(model.wv[word], model.wv[bad_word])])
        except:
            B[bad_word].extend([float("inf")])
```

```python
t = 0.5
```

```python
C_i = 0
best = None
d = float("inf")
for i in range(1, len(red_words)+1):
    for rc in itertools.combinations(red_words, i):
        for word in range(len(words)):
            wd = float("inf")
            for bad_word in bad_words:
                if B[bad_word][word] < wd:
                    wd = B[bad_word][word]
                dr = 0
                for red_word in rc:
                    if R[red_word][word] > dr:
                        dr = R[red_word][word]
                    if dr < d and dr < wd and dr < t:
                        d = dr
                        best = words[word]
                        C_i = i
```
